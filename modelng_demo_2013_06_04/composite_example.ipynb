```python
import numpy as np
from modeling import *
```

Create an ellipse and a circle with a common origin.

```python
def ellipse(r1, r2):
    alpha = np.arange(0, 2* np.pi, .1)
    x = r1 * np.sin(alpha)
    y = r2 * np.cos(alpha)
    return x, y

x, y = ellipse(2, 2)
ex, ey = ellipse(4, 2)
plt.plot(x, y)
plt.plot(ex, ey)
```

Shift the circle in x, y.

```python
offx = models.ShiftModel(-1.3)
offy = models.ShiftModel(-2.1)
x1 = offx(x)
y1 = offy(y)
plt.plot(x1, y1)
plt.plot(ex, ey)
```

Rotate the ellipse

```python
rot = models.MatrixRotation2D(angle=90)
ex1, ey1 = rot(ex, ey)
plt.plot(x1, y1)
plt.plot(ex1, ey1)
```

Apply a stereograpfic projection to the ellipse.

```python
stg=models.Pix2Sky_STG()
ex2, ey2 = stg(ex1,ey1)
plt.plot(x1, y1)
plt.plot(ex2, ey2)
```

All these steps can be done using a composite model. The input is passed as a `LabeledInput` which is simply a dictionary mapping inputs and labels. The 1inmap` and `outmap` parameters show the input and output of each transform.

```python
scomp = SCompositeModel([offx, offy, rot, stg], inmap=[['x'], ['y'], ['ex', 'ey'], ['ex', 'ey']], outmap=[['x'], ['y'], ['ex', 'ey'], ['ex', 'ey']])

labeled_input = LabeledInput([ex, ey, x, y], ['ex', 'ey', 'x', 'y'])
result = scomp(labeled_input)
plt.plot(result.x, result.y)
plt.plot(result.ex, result.ey)
```

If all models in the composite have implement inverse(), getting the inverse of an `SCompositeModel` is simple:

```python
inv_scomp = scomp.inverse()
new_result = inv_scomp(result)
```

```python
plt.plot(new_result.x, new_result.y)
plt.plot(new_result.ex, new_result.ey)
```
